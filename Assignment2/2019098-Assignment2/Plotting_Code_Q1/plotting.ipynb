```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('final_ques_1.csv')
```

```python
cumulativeSum = data.cumsum()
cumulativeSum = cumulativeSum['Length']
```

```python
plot_points_y_cont = []
plot_points_x_cont = []

for i in range(0, 60):
    range_indices = data.loc[(data['Time'] >= (2*i)) & (data['Time'] < (2*(i + 1)))].index
    start_point, end_point = range_indices[0], range_indices[len(range_indices) - 1]
    num_bytes = cumulativeSum[end_point] - cumulativeSum[start_point] + data.loc[start_point]['Length']
    end_time = data.loc[end_point]['Time']
    start_time = data.loc[start_point]['Time']

    throughput = float(num_bytes) / float(end_time - start_time)

    plot_points_y_cont.append(throughput)
    plot_points_x_cont.append(2*(i + 1))

plt.plot(plot_points_x_cont, plot_points_y_cont)
plt.xlabel('Time')
plt.ylabel('Throughput')
plt.show()
```
